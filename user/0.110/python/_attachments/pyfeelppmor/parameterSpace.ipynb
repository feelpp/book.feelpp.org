# ParameterSpace

Class `ParameterSpace` (more precisely `feelpp.mor._mor.ParameterSpace`). Space of the parameters of the problem.


```python
Dmu = model.parameterSpace()
```

## Methods


- `element(broadcast: bool = True, apply_log: bool = False) &#8594; feelpp.mor._mor.ParameterSpaceElement` : return a parameter from the space :



- `broadcast` : share the parameter to all processors


- `apply_log` : log random chosen parameter



- `sampling() &#8594; feelpp.mor._mor.ParameterSpaceSampling`


- `mumin() &#8594; feelpp.mor._mor.ParameterSpaceElement` : return an element with the minimal values.


- `mumax() &#8594; feelpp.mor._mor.ParameterSpaceElement` : return an element with the maximal values.
fpp


- `dimension() &#8594; int` : return the dimension of the space


- `parameterNamed(name: str)` &#8594; `float` : return the parameter named `str`


- `parameterNames(name: str)` &#8594; `list` : return the list of the name of the parameters

## Static method


- `New(feelpp._models.ModelParameters, arg1: feelpp._core.WorldComm)`: constructor using a model parameters loaded from a JSON file (see [ModelParameters](../pyfeelpptoolboxes/modelproperties.ipynb)).


```python
modelParameters = heatBox.modelProperties().parameters()    # for instance
Dmu = feelpp.mor._mor.ParameterSpace.New(modelParameters, feelpp.Environment.worldCommPtr())
```


- `create(arg0: int) &#8594; feelpp.mor._mor.ParameterSpace` : return a `ParameterSpace` of a dimension `arg0`.
