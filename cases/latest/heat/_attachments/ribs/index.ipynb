# Bracing with ribs

## Introduction

The ribs have stiffening effect. So the plates are supplied with different ribs.


## Running the case

The command line to run this case is
```mpirunmpirun -np 4 feelpp_toolbox_solid --case "github:{repo:toolbox,path:examples/modules/csm/examples/ribs}"```*Note:* The report of the execution of the command above is available .

## Materials and boundary conditions

## Materials

The beam is made of steel, so we need to use the following data:

- E = 210 GPa &#8594; 210e3 N/mm2
- nu = 0.3
- rho = 7800 kg/m3 &#8594; 7800e-9 kg/mm3
